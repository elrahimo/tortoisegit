2.8 Moving average 

```python
print("Input")
np.random.seed(100)
a = np.random.randint(10, size=10)
np.convolve(a, np.ones(3)/3, mode='valid')

def moving_average(a, n=3) :
    ret = np.cumsum(a, dtype=float)
    ret[n:] = ret[n:] - ret[:-n]
    return ret[n - 1:] / n

moving_average(a)
#Berechnung des gleitenden Durchschnitts der Fenstergröße 3 für das gegebene 1D-Array:
```

```python
def first_order_difference(f,x0):
    h=0.000000001
    f1 = (f(x0 + h) - f(x0 - h))/(2*h)
    return f1
```

```python
def testf(x):
    return pow(x,4)


def unit_test():
    x0 = 1
    np.testing.assert_almost_equal(first_order_difference(testf, x0), 4)
  

```

```python
first_order_difference(testf, 1.0)
unit_test()
def f(x):

    k=x**4-3*x**2-x-2
    return k
first_order_difference(f,1)

first_order_difference(testf, 1.0)
```

```python
def gradient_descent(learning_rate,x,f,iteration=500):
    
    for i in range(iteration):
        x=x-learning_rate*first_order_difference(f,x)
    
        
    return x
```

```python
def testf(x):
    return x**2
print(gradient_descent(0.01,-2,testf))
```
